```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
# Diffusion function
g = lambda s: 1/(1+s)

# Test function
def f(x, alpha):
    first = 5-np.tanh(alpha*(x-1))-np.tanh(alpha*(x-2))
    second = np.tanh(alpha*(x-4)) + np.tanh(alpha*(x-5)) + 0.1*(np.sin(5*x))**2*np.sin(50*x)
    return first+second

f = lambda x, α: 5 - np.tanh(α*(x-1)) - np.tanh(α*(x-2)) \
    + np.tanh(α*(x-4)) + np.tanh(α*(x-5)) \
    + 0.1*np.sin(x)**2 * np.sin(50*x)
```

```python
# Create random 1D function
np.random.seed(1338)
I = np.zeros(M+2)
s = (M + 2)//5
for i in range(4):
    I[i*s:(i+1)*s] = np.random.randint(0, 5)
I = I + np.random.normal(0, 0.2, size = M+2)
plt.plot(I)
plt.show()
```

```python
# Number of points in x- and y-direction respectively
M = 400
N = 50

#Spatial steps
dx = 1/(M+1)
dy = 1/(N+1)

# Number of iterations, and timestep
T = int(1e4)
dt = dx**2

# Initiate grid
U = np.zeros((T, M+2))
U[0] = f(np.linspace(0, 6, M+2), 30)
# U[0] = I
```

```python
# Forward difference matrix(with backward for last)
Dx = -1 * np.eye(M+2,k = -1) + 1 * np.eye(M+2, k = 1)
Dx[0,:] = 0
Dx[-1,:] = 0
Dx /= dx

# Matrices to help construct A-matrix
Γ = np.eye(M+2, k = 0) + np.eye(M+2, k = 1)
Γ[0,:3] = [0, 1/2, 1/2]
Γ[-1,-2:] = [-1/2, -1/2]

Ω = (-1) * np.eye(M+2, k = -1) + (-2)*np.eye(M+2, k = 0) + (-1) * np.eye(M+2, k = 1)
Ω[0,:3] = [1/2, 0, -1/2]
Ω[-1,-3:] = [-1/2, 0, 1/2]

Ξ = np.eye(M+2, k = 0) + np.eye(M+2, k = 1)
Ξ[0, :2] = [-1/2, -1/2]
Ξ[-1, -2:] = [1/2, 1/2]
```

```python
A = np.zeros((M+2, M+2))
for it in range(T-1):
    G = g(Dx.dot(U[it])**2)
    γ = Γ.dot(U[it])
    ω = Ω.dot(U[it])
    ξ = Ξ.dot(U[it])
    A[0, :3] = [ξ[0], ω[0], γ[0]]
    A[-1, -3:] = [ξ[-1], ω[-1], γ[-1]]
    for i in range(1, M+1):
        A[i, i-1:i+2] = [ξ[i], ω[i], γ[i]]
    U[it+1] = (np.identity(M+2) + dt * A).dot(U[it])
```

```python
plt.figure()
plt.plot(U[0])

plt.figure()
plt.plot(U[T//2])

plt.figure()
plt.plot(U[-1])
plt.show()
```

```python
# Number of points in x- and y-direction respectively
M = 400
N = 50

#Spatial steps
dx = 1/(M+1)
dy = 1/(N+1)

# Number of iterations, and timestep
T = 100
dt = 0.01

# Initiate grid
U = np.zeros((T, M+2))
# U[0] = f(np.linspace(0, 6, M+2), 30)
U[0] = I
```

```python
A = np.zeros((M+2, M+2))
for it in range(T-1):
    G = g(Dx.dot(U[it])**2)
    γ = Γ.dot(U[it])
    ω = Ω.dot(U[it])
    ξ = Ξ.dot(U[it])
    A[0, :3] = [ξ[0], ω[0], γ[0]]
    A[-1, -3:] = [ξ[-1], ω[-1], γ[-1]]
    for i in range(1, M+1):
        A[i, i-1:i+2] = [ξ[i], ω[i], γ[i]]
    U[it+1] = np.linalg.solve((np.identity(M+2) - dt * A), U[it])
    
    
plt.figure()
plt.plot(U[0])

plt.figure()
plt.plot(U[T//2])

plt.figure()
plt.plot(U[-1])
plt.show()
```
